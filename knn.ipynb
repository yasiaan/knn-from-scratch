<strong>2 - Créez une fonction qui crée une liste L de tous les mots figurant dans cette base (sans
répétition).</strong>

```python
# Creation de la liste L de tous les mots figurant dans cette base (sans répétition) et la stocker dans la variable vocab.
def creationL(liste_s):
    words = []
    for s in liste_s:
        w = extraction(s)
        words.extend(w)
    words = sorted(list(set(words)))
    return words

def extraction(s):
    punctuation= '''!()-–[]{};:'"\, <>./?@#$%^&*_~«»'''
    words = re.sub("^[a-zA-Z]+", " ",  s).split()
    cleaned_text = []
    for w in words:
        w_new = ''
        for x in w:
            if x not in punctuation and not x.isnumeric():
                w_new = w_new + x
        if len(w_new) == 0:
            continue
        cleaned_text.append(w_new.lower())
    return cleaned_text  


vocab = creationL(t)
old_len_vocab = len(vocab)
print("{0} mots dans cette liste pour les '15 textes'\n \n{1} \n".format(len(vocab), vocab))
```

<strong>3 - Créez une première fonction T2V qui transforme chaque texte T en un vecteur
d’occurrences des mots de la liste L.</strong>

```python
def vector( v, s):
    words = extraction(s)
    bag_vector = numpy.zeros(len(v))
    for w in words:
        for i,word in enumerate(v):
            if word == w: 
                bag_vector[i] += 1
    return list(bag_vector)

def t2v(alls, v):        
    bags = []
    for s in alls:
        bag_vector = vector(v, s)
        print("\n{0}\n{1}".format(s,numpy.array(bag_vector)))
        bags.append(bag_vector)
    return bags
bg = t2v(t, vocab)
```

```python
df_t2v = pd.DataFrame(bg, columns=vocab)
df_t2v
```

<strong>4 - Créez une liste de mots vides MV (pronoms, prépositions, conjonctions) et utilisez
la pour réduite la dimension de L. Quel est le taux de réduction obtenu ?</strong>

```python
def mv(vc):
    ignore = get_stop_words('french')
    v = vc
    for x in v:
        if x in ignore:
            v = [temp for temp in v if temp != x ]
    return v
mv_vocab = mv(vocab)
mv_len_vocab = len(mv_vocab)
print("Le taux de reduction est de : {0} %\n".format(1-(mv_len_vocab/old_len_vocab)))
print("{0} words in this List for the '15 textes' \n \n {1} \n".format(len(mv_vocab), mv_vocab))
```

```python
bg_mv = t2v(t, mv_vocab)
```

```python
df_mv = pd.DataFrame(bg_mv, columns=mv_vocab)
df_mv
```

<strong>6 - Créez une fonction DIST qui calcule la distance euclidienne entre deux vecteurs</strong>

```python
def dist( a, b):
    result = 0
    if len(a) == len(b):
        for i in range(len(a)):
            result += (a[i]-b[i])**2
        result = math.sqrt(result)
    else:
        print("les deux vecteurs n'ont pas la même longueur !!")
    return result
```

<strong>7 - Proposez un texte pour lequel : (en utilisant T2V)
- Parmi les 15 textes de la base, la DIST minimale correspond à la
classe économie.</strong>

```python
#définition de KNN
def knn(k,t,B):
    #créer la table des distances t-B
    B = [[b,t] for b,t in zip(B,tps)]
    tD=[]
    for i in range(len(B)):
        tD.append([dist(t,B[i][0]),B[i][1]])
    #trier la tD
    for i in range(len(tD)):
        for j in range(len(tD)):
            if tD[i][0]>tD[j][0]:
                temp=tD[i]
                tD[i]=tD[j]
                tD[j]=temp
    pp.pprint(tD)
    #calculer les k derniers de la tD
    #créer les tables tC et tO
    tC=[]
    tO=[]
    for i in range(len(tD)):
        if not B[i][1] in tC:
            tC.append(B[i][1])
            tO.append(0)
    
    #remplir la table tO
    for i in range(1,k+1):
        for j in range(len(tC)):
            if tD[-i][1]==tC[j]:
                tO[j]=tO[j]+1
    pp.pprint(tO)
    #retourner la classe prédite
    if tO.count(max(tO)) == 1:
        indmax=tO.index(max(tO))
        return tC[indmax]
    return tD[-1][1]
```

```python
texte_eco = "Les exportations de l'économie grecque sont traditionnellement faibles."
bag_vector_eco = vector(vocab, texte_eco)

predicted_eco = knn(4, bag_vector_eco, bg)
print("\nParmi les 15 textes de la base, la DIST minimale (en utilisant T2V) correspond à la classe : {0}".format(predicted_eco.upper()))
```
